```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv('Info_pais.csv', encoding = 'ISO-8859-1', delimiter = ";")
```

```python
df.head()
```

```python
df_order = df.sort_values('Esperanza de vida',ascending=True)
```

```python
df_order.head()
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'])
plt.title('Renta per capita vs Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')
plt.yticks(ticks=range(0, 100, 5))
plt.tight_layout()
plt.show()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / max(df_order['Poblacion'])
```

```python
df_order.head()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / (max(df_order['Poblacion'])/ 10000)
```

```python
df_order.head()
```

```python
plt.figure(figsize=(10, 8))  # Ajustar tamaño de la figura
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'], s=df_order['Poblacion_normalizada'])
plt.title('Renta per capita vs Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')
plt.yticks(ticks=range(0, 85, 5))  # Establecer ticks del eje y
plt.tight_layout()
plt.show()
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per capita vs Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')
plt.yticks(ticks=range(0, 85, 5))  # Establecer ticks del eje y
fig = plt.gcf()
fig.set_size_inches(14.5,10);
#fig de figura
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per capita vs Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')
plt.yticks(ticks=range(0, 85, 5))  # Establecer ticks del eje y
fig = plt.gcf()
fig.set_size_inches(14.5,10);
#fig de figura

for i in range(0,10):
    plt.annotate(
        df_order['País'][i],
        (
            df_order['Renta per capita'][i],
            df_order['Esperanza de vida'][i]
        )
    ) 
```

##Con el resultado de el grafico podemos apreciar que de cierta manera las personas que tienen un mayor ingreso economico tienen una esperanza de vida mayor, quiza por que el acceso a ciertos recursos les permite mejorar su calidad de vida, pero por otro lado veo que italia y japon sus ingresos no son muy altos y tienen mayor esperanza de vida que estados unidos por ejemplo. Tambien hay cierta tendencia de que las personas que tienen pocos ingresos sus esperanza de vida es menor.

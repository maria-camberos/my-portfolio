```python
import sys 
import openpyxl
import matplotlib.pyplot as plt
import pandas as pd
from openpyxl import Workbook
from openpyxl import load_workbook
import os, stat

def solitario():
    print("MODO SOLITARIO")
    import random
    numero_r = random.randrange(1,100)
    nombre_uno = input("Hola, ¿Cual es tu nombre?: ")    
    print("Selecciona un numero del 1 al 3 para el grado de dificultad: \n1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)")    
    print("                ")
    print(numero_r)
   
    contador = 0
    dificultad = int(input("Ingresa el grado de dificultad: "))
    print("                ")

    while True:
    
        if dificultad == 1:
            intentos = 20
        elif dificultad == 2:
            intentos = 12
        else: 
            intentos = 5
            
        num_adivinar = int(input("Ingresa un numero a adivinar: "))
        if num_adivinar > numero_r:
            contador = contador +1
            if contador == intentos:
                print(".......................¡EXCEDISTE EN NUMERO DE INTENTOS!.......................")
                break
            print("El numero a adivinar es menor, intenta de nuevo: ")
        elif num_adivinar < numero_r:
            contador = contador +1
            if contador == intentos:
                print(".......................¡EXCEDISTE EN NUMERO DE INTENTOS!.......................")
                break
                print("Contador", contador)
            print("El numero a adivinar es mayor, intenta de nuevo: ")
        else:
            print("¡FELICIDADES ACERTASTE EL NUMERO")
            print("                               ")          
            break
        
    nombre = []
    no_intentos = []
        
    nombre.append(nombre_uno)
    no_intentos.append(intentos)
        
    data = {'Nombre': nombre,
                'Intentos':no_intentos}
    
    #Guardar el informacion del jugador    
    df = pd.DataFrame(data)
    file_path = 'EstadisticasX2.xlsx'
    saveplayer(file_path,df)  #Aqui se llama la funcion Para guardar la opcion solitario
    print("=========================================================================================")

            
def dos_jugadores():    
    print("MODO DOS JUGADORES")
    jugador_nombre1 = input("Hola, Ingresa el nombre del primer jugador: ")
    jugador_nombre2 = input("Hola, Ingresa el nombre del segundo jugador: ")
    num_jugador1 = int(input("Ingresa el numero para ser adivinado por el segundo jugador: "))
        
    print("Selecciona un numero del 1 al 3 para el grado de dificultad: \n1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)")    
    print("                ")   
    contador = 0
    dificultad = int(input("Ingresa el grado de dificultad: ")) 
    print("                ")

    while True:    
        if dificultad == 1:
            intentos = 20
        elif dificultad == 2:
            intentos = 12
        else: 
            intentos = 5
            
        num_jugador2 = int(input("Ingresa un numero a adivinar: "))
        if num_jugador1 > num_jugador2:
            contador = contador +1
            if contador == intentos:
                print(".......................¡EXCEDISTE EN NUMERO DE INTENTOS!.......................")
                break
            print("El numero a adivinar es mayor, intenta de nuevo: ")
        elif num_jugador1 < num_jugador2:
            contador = contador +1
            if contador == intentos:
                print(".......................¡EXCEDISTE EN NUMERO DE INTENTOS!.......................")
               
                break
                print("Contador", contador)
            print("El numero a adivinar es menor, intenta de nuevo: ")
        else:
            print("¡FELICIDADES ACERTASTE EL NUMERO!")
      
            print("                               ")
            break
       
    print("=========================================================================================")
           

def estadistica():
    print("ESTADISTICAS")
    excel_doc = openpyxl.load_workbook("EstadisticasX2.xlsx")
    Hoja1 = excel_doc["Sheet1"]
       
    
    workbook1 = "EstadisticasX2.xlsx"
    print(".......................RESULTADOS.......................")
    df = pd.read_excel(workbook1)
    valores = df[['Nombre', 'Intentos']]
    graf = valores.plot.bar(x='Nombre', y ='Intentos', rot=0)
    plt.show()
    
    print("                        ")
    print("     Estadisticas de otros jugadores")
    x = ['Juan Diaz', 'Ricardo Lopez', 'Ana Miranda', 'Mario Vaca', 'Soledad Rios']
    y = [6,12,3,8,15]
    plt.bar(x,y)
    plt.show()
    print(".......................................................")
    print("                ")
    
    print("=========================================================================================")
    
def salir():
    
    print(".......................¡GRACIAS POR JUGAR!.......................")
    
    print("                ")
    sys.exit("[][][][][][][][][][][][][]  ¡JUEGO TERMINADO! [][][][][][][][][][][][][]") 
    print("=========================================================================================")

    
# Verifica la existencia del archivo con el metodo os.path.exists .
def saveplayer(file_path,datap):
    df = pd.DataFrame(datap)
    ret = False
    # Si el archivo existe.
    if(os.path.exists(file_path)):
        ret = True
        print(file_path + " log existe y sera actualizado.")
        book = load_workbook('EstadisticasX2.xlsx')
        writer = pd.ExcelWriter('EstadisticasX2.xlsx', engine='openpyxl')
        writer.book = book
        writer.sheets = {ws.title: ws for ws in book.worksheets}
        for sheetname in writer.sheets:
            df.to_excel(writer,sheet_name=sheetname, startrow=writer.sheets[sheetname].max_row, index = False,header= False)

        writer.save()        
    #si no existe el archivo
    else:
        ret = False
        print(file_path + " log no existe y sera creado")
        filename = 'EstadisticasX2.xlsx'
        writer = pd.ExcelWriter('EstadisticasX2.xlsx', engine='xlsxwriter')
        df.to_excel(writer, sheet_name='Sheet1', index=False)
        writer.save()
        
    return ret
# Aqui Termina El proceso para guardar los jugadores
    
    
def Menu():
    
    while True:
        print("MENU PRINCIPAL")
        print("               ")
        print("¡ADIVINA EL NUMERO!")
        print("Adivina un numero entre 1 y 100 ")
        print("Selecciona una opcion del 1 al 4 para comenzar: ")    
        print("1. Partida modo solitario\n2. Partida 2 Jugadores\n3. Estadística\n4. Salir")
        modo_juego = int(input("Ingresa una opcion del uno al 4 para continuar:  "))        
        print("                ")
        print("=========================================================================================")

        if modo_juego == 1:
            solitario()
        elif modo_juego == 2:
            dos_jugadores()
        elif modo_juego == 3:
            estadistica()
        elif modo_juego == 4:
            salir()
        else:
            print('===========INTRODUCE UNA OPCION VALIDA (1-4) ==============')
        print("                ")
       
        
Menu()
```
